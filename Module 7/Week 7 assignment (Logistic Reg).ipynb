```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 20, 10
import pandas as pd
import numpy as np
import os

from sklearn.linear_model import LogisticRegression
from pylab import rcParams
rcParams['figure.figsize'] = 20, 10

from sklearn.tree import DecisionTreeClassifier

from sklearn.linear_model import LogisticRegression as Model
from sklearn.metrics import (accuracy_score, 
                             classification_report, 
                             confusion_matrix, auc, roc_curve
                            )
from sklearn import linear_model, metrics
from sklearn.model_selection import train_test_split

#bringing in data 

filename = "/Users/sanmirkem/Desktop/mlnn-personal/data/Heart.csv"
heart_data = pd.read_csv(filename)
heart_data
```

```python
heart_data['AHD']=heart_data['AHD'].map({'Yes':1,'No':0})
heart_data.head()
```

```python
heart_data.drop(heart_data.columns[0], axis = 1, inplace = True)
heart_data.head()
```

```python
heart_data.drop(columns=['ChestPain','Thal'], axis=1,inplace=True)
heart_data.head()
```

```python
HD = heart_data.AHD.value_counts()[0]
No_HD = heart_data.AHD.value_counts()[1]
print('Heart Disease: ', HD)
print('No Heart Disease: ', No_HD)
```

```python
#checking for NAs
heart_data.isnull().sum()
heart_data = heart_data.dropna(axis=0,subset=['Ca'])
```

```python
heart_data.isnull().sum()
```

```python
# Splitting data into testing and training 
x = heart_data.drop(['AHD'], axis = 1)
y = heart_data.AHD.values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# conducting logistic regression 

heart_data.replace([np.inf, -np.inf], np.nan, inplace=True)
logistic_regression_model = LogisticRegression()
logistic_regression_model.fit(x_train, y_train) 
logistic_predictions =logistic_regression_model.predict(x_test)
logistic_regression_accuracy=accuracy_score(y_test, logistic_predictions)*100


print('Logistic Regression Testing Accuracy: ',logistic_regression_accuracy)
```

```python
print(classification_report(y_test, logistic_predictions))
```

```python
confusion_matrix(y_test, logistic_predictions)
```

```python
clf = DecisionTreeClassifier(criterion='entropy') 
clf.fit(x_train, y_train)

# making predictions
clf_pred = clf.predict(x_test) # Test the prediction accurracy of the model
accuracy_score(y_test, clf_pred)
```

```python
print(classification_report(y_test, clf_pred))
```

```python
confusion_matrix(y_test, clf_pred)
#After running our models, we see that the decision tree outperformed the logistic regression in some areas. 
# For e.g., our confusion matrix shows the decision tree predicted more1's compared to the
                                #logisitic model.
# Despite having a lower recall percentage of 0's compared to the Logistic model
                #the decision had a smaller disparity in f1-score between the health disease diagnosis (AHD). 
    #I assume this is a good thing?
```

```python
clf_v2 = DecisionTreeClassifier(max_depth=11) 
clf_v2.fit(x_train, y_train)

# making predictions
clf_pred_v2 = clf_v2.predict(x_test) # Test the prediction accurracy of the model
accuracy_score(y_test, clf_pred_v2)
```

```python
print(classification_report(y_test, clf_pred_v2))
```

```python
confusion_matrix(y_test, clf_pred_v2)
```
